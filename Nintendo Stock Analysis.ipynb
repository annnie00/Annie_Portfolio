# Importing new data for comparison and more plots!

```python
file_path = "/Users/annievernaza/Downloads/vgsales.csv"
nintendo = pd.read_csv(file_path)
```

```python
nintendo.head()
```

```python
nintendo = nintendo[(nintendo['Publisher'] == 'Nintendo') & (nintendo['Year'] >= 2005) & (nintendo['Year'] <= 2015)]
```

```python
nintendo.head()
```

Changing column Data Type
- Year column from float to int

```python
nintendo_summary = nintendo.groupby('Year').agg({'Name': 'count', 'Global_Sales': 'sum'}).reset_index()
nintendo_summary.columns = ['Year', 'Number_of_Nintendo_Games', 'Sum_Global_Sales']
print(nintendo_summary)
```

```python
nintendo_summary=nintendo_summary.astype({"Year": int})
```

```python
nintendo_summary.dtypes
```

```python
print(nintendo_summary)
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.plot(nintendo_summary['Year'], nintendo_summary['Sum_Global_Sales'], linewidth=1)
plt.xlabel('Year')
plt.ylabel('Sum_Global_Sales')
plt.title('Nintendo Global Sales 2005-2015')
plt.grid(True)
plt.show()

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.xlim(pd.Timestamp('2005-01-01'), pd.Timestamp('2015-12-31'))

plt.plot(df['Date'], df['Close'], linewidth=1)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Nintendo Stock Price Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
total_sales = nintendo.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()
print(total_sales)
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.plot(total_sales.index, total_sales['NA_Sales'], label='NA Sales')
plt.plot(total_sales.index, total_sales['EU_Sales'], label='EU Sales')
plt.plot(total_sales.index, total_sales['JP_Sales'], label='JP Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales (Millions)')
plt.title('Nintendo Total Sales Per Year')
plt.legend()
plt.grid(True)
plt.show()
```
